# Vectors

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#sets default grids and other things to our graph
sns.set()
```

```python
#Plotting a vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#plotting mulptile vectors
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#addition of two vectors
vec1 = np.asarray([0,0,2,3])
vec2 = np.asarray([0,0,3,-2])
add = vec1 + vec2
print(add)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#multiplying the vector with a scalar
vec1 = np.asarray([0,0,2,3])
vec2 = -0.5 * vec1
print(vec2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-1,-1.5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#dot product of two vectors
a = np.array([2,3])
b = np.array([4,4])
dot_prod = np.dot(a,b)
```

```python
print(dot_prod)
```

```python
c = np.array([40,20,35])
d = np.array([53,24,68])
c_dot_d = np.dot(c,d)
```

```python
print(c_dot_d)
```

```python
#Cross Product of two vectors
c = np.array([40,20,35])
d = np.array([53,24,68])
cross_prod = np.cross(c,d)
```

```python
print(cross_prod)
```

```python
#projection of a vector on b vector
a = np.array([23,45,62])
b = np.array([45,82,67])
mag_b = np.sqrt(sum(b**2))
proj = (np.dot(a,b)/mag_b**2)*b
print('Projection of a on b vector: ',proj)
```

```python
#creating matrices with random values
random_matrix = np.random.rand(3,3)
print(random_matrix)
```

```python
int_matrix = np.random.randint(100,size=(4,5))
print(int_matrix)
```

```python
ones_matrix = np.ones((3,3),dtype=int)
print(ones_matrix)
```

```python
#creating null matrix
zeroes_matrix = np.zeros((3,3))
print(zeroes_matrix)
```

```python
#Identity Matrix
identity_matrix = np.eye(4,4)
print(identity_matrix)
```

```python
#transpose of a matrix
print("Before Transpose: \n",int_matrix)
print("\nAfter Transpose: \n",int_matrix.T)
```

```python
#matrix addition
matrix1 = np.random.randint(100,size=[3,3])
matrix2 = np.random.randint(100,size=[3,3])
print(matrix1,'\n')
print(matrix2)
```

```python
#matrix addition
#np.sum(matrix1,matrix2)
add = matrix1 + matrix2
print(add)
```

```python
#matrix subtraction
#np.subtract(matrix1,matrix2)
sub = matrix1 - matrix2
print(sub)
```

```python
#multiplying scaler to the matrix
x = 5
y = np.random.randint(10,size=[5,5])
scalar_mult = x*y
print(f'{y}\n\nAfter multiplying by {x}:')
print(scalar_mult)
```

```python
#multiplying two matrices
matrix1 = np.random.randint(5,size=[3,3])
matrix2 = np.random.randint(5,size=[3,4])
print(matrix1,'\n')
print(matrix2)
```

```python
product = np.dot(matrix1,matrix2)
print(product)
```

```python
#element wise matrix multiplication
matrix1 = np.random.randint(5,size=[3,3])
matrix2 = np.random.randint(5,size=[3,3])
print(matrix1,'\n')
print(matrix2,'\n')
prod = np.multiply(matrix1,matrix2)
print(prod)
```
