```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_df = pd.read_csv('mail_data.csv')
mail_df.head()
```

```python
mail_df.shape
```

```python
mail_df['Category'].value_counts()
```

```python
mail_df.isnull().sum()
```

```python
#if any na values where there the below code can help replace it with empty string
mail_df = mail_df.where(pd.notnull(mail_df),'')
```

```python
mail_df.head()
```

```python
#label encoding spam as 1 and ham as 0
mail_df.loc[mail_df['Category'] == 'spam','Category',] = 1
mail_df.loc[mail_df['Category'] == 'ham','Category',] = 0
```

```python
X = mail_df['Message']
Y = mail_df['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
#train test split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#feature extraction ie transform the text data to feature vectors that can be used as input to Logistic reg model
feature_extraction = TfidfVectorizer(min_df = 1,stop_words='english',lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

#convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
#training logistic regression model
model = LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
#evaluating the model
training_pred = model.predict(X_train_features)
training_accuracy = accuracy_score(Y_train,training_pred)
```

```python
print("Accuracy on training data: ",training_accuracy)
```

```python
test_pred = model.predict(X_test_features)
test_accuracy = accuracy_score(Y_test,test_pred)
```

```python
print("Accuracy on training data: ",test_accuracy)
```

```python
input_msg = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's "]
input_features = feature_extraction.transform(input_msg)

#making predictions
prediction = model.predict(input_features)

if(prediction[0]==0):
    print("The mail is not Spam!")
else:
    print("The mail is Spam!")
```
