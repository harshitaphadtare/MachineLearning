```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
import kagglehub

path = kagglehub.dataset_download("nehalbirla/vehicle-dataset-from-cardekho")

print("Path to dataset files:", path)
```

```python
import os

print("Files in dataset folder:", os.listdir(path))
```

```python
file_path = os.path.join(path, "car data.csv")  
df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Fuel_Type'].value_counts()
```

```python
df['Seller_Type'].value_counts()
```

```python
df['Transmission'].value_counts()
```

```python
#label encoding Fuel Type Column
df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

#label encoding Seller Type Column
df.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

#label encoding Transmission Type Column
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Car_Name','Selling_Price'],axis=1)
Y = df['Selling_Price']
```

```python
print(X)
```

```python
print(Y) 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#loading the linear regression model
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,Y_train)
```

```python
Y_pred = lin_reg.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train,Y_pred)
print('R squared error of training data: ',error_score)
```

```python
plt.scatter(Y_train,Y_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
Y_pred_test = lin_reg.predict(X_test)
```

```python
error_score_test = metrics.r2_score(Y_test,Y_pred_test)
print('R squared error of test data: ',error_score_test)
```

```python
plt.scatter(Y_test,Y_pred_test)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
lasso_model = Lasso()
```

```python
lasso_model.fit(X_train,Y_train)
```

```python
Y_train_pred = lasso_model.predict(X_train)
```

```python
error_score_train = metrics.r2_score(Y_train,Y_train_pred)
print('R squared error: ',error_score_train)
```

```python
plt.scatter(Y_train,Y_train_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
Y_test_pred = lasso_model.predict(X_test)
```

```python
error_score_test = metrics.r2_score(Y_test,Y_test_pred)
print('R squared error: ',error_score_test)
```

```python
plt.scatter(Y_test,Y_test_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
