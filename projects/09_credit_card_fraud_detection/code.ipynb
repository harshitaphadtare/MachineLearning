```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

```python
import os
print('Files in path: ',os.listdir(path))
```

```python
file_path = os.path.join(path,'creditcard.csv')
df = pd.read_csv(file_path)
df.head()
```

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
#compare values for both transactions
df.groupby('Class').mean()
```

```python
#Under-Sampling
#Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions
legit_sample = legit.sample(n=492)
```

```python
#concatinating two dataframes
new_df = pd.concat([legit_sample,fraud],axis=0)
new_df.head()
```

```python
new_df.loc[23093]
```

```python
new_df.shape
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X = new_df.drop(columns='Class',axis=1)
Y = new_df['Class']
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify=Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model training
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_train_pred = model.predict(X_train)
Y_train_accuracy = accuracy_score(Y_train_pred,Y_train)
print("Accuracy Score of training data ",Y_train_accuracy)
```

```python
Y_test_pred = model.predict(X_test)
Y_test_accuracy = accuracy_score(Y_test_pred,Y_test)
print("Accuracy Score of test data ",Y_test_accuracy)
```
